```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set_theme()
```

```python
df = pd.read_csv("../data/real_estate_price_size.csv")
```

```python
df.head()
```

```python
x1 = df["size"]
y = df["price"]
```

```python
plt.scatter(x1, y)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary() 
```

```python
plt.scatter(x1, y)
yhat = x1 * 223.1787 + 101900
fig = plt.plot(x1, yhat, lw=2, c="orange")
plt.xlabel("Size")
plt.ylabel("Price")
plt.show()
```
