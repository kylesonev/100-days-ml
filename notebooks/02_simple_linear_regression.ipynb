```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../data/student_scores.csv")
df
```

```python
df.head(3)
df.isnull().sum()
```

```python
X = df.iloc[:, 0].values
y = df.iloc[:, 1].values
```

```python
X_train, X_test,y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
```

```python
regressor = LinearRegression()
regressor = regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color="red", label="Historical data")
plt.plot(X_test, y_pred, color="blue", label="Predict(Regression Line)")


plt.title("Hours of Study vs Score")
plt.xlabel("Hours")
plt.ylabel("Score")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import seaborn as sns 
df_results = pd.DataFrame({
    'Hours': X_test.flatten(),
    'Historical': y_test,
    'Predicted': y_pred
})

df_results = df_results.sort_values(by='Hours')

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hours', y='Historical', data=df_results, color='red', label='Valores reais')
sns.lineplot(x='Hours', y='Predicted', data=df_results, color='blue', label='Previsão')

plt.title('Horas de Estudo vs Pontuação (Previsão)')
plt.xlabel('Horas de Estudo')
plt.ylabel('Pontuação')
plt.legend()
plt.grid(True)
plt.show()
```
